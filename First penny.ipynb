```python
#My first pennylane code

import pennylane as pen
from pennylane import numpy as np

dev = pen.device("default.qubit", wires = 0)


@pen.qnode(dev)

def circuit(x):
    pen.RX(x, wires = 0)

    return pen.expval(pen.PauliZ(0))

angle = np.pi/2 + 3

result = circuit(angle)

print(f" Output expectation value: {result:.1f}")




```
